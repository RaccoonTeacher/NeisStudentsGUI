```python
import tkinter as tk # tkinter를 사용하겠다
import tkinter.messagebox # tkinter의 messagebox의 사용을 위해 불러와야 함
import tkinter.font
import re # 정규표현식을 위함
import os # 파일 존재 유무를 알아보기 위함
import openpyxl as xl
import time
import pyautogui as pag
import keyboard
import pyperclip
import sys

root = tk.Tk()
menubar = tk.Menu(root) # 메뉴바 추가

def execute_toplevel(): # 외부 윈도우 생성
    toplevel = tk.Toplevel(root, menu = menubar)
    toplevel.geometry("500x300+300+140")
        #설명란 추가
    toplevel_label = tk.Label(toplevel, text="학생 이름을 입력하세요.\n학생 이름 사이에는 \'쉼표\' 혹은 \'띄어쓰기\'가 되어 있어야 합니다.\n모두 입력하였다면 아래의 \'클릭\'버튼을 누르세요.")
    toplevel_label.pack()
        #Entry에 학생 이름 입력하고 그 값을 메모장에 저장
    toplevel_entry_type = tk.Entry(toplevel, width=200, background="#99FFFF") #Entry 생성
    toplevel_entry_type.pack() #엔트리 보이게
    def memo_write(event=None): # 메모장에 학생 이름 넣기 위한 함수 생성
        File = open("Students Name.txt", "w", encoding = "UTF-8") # 메모장 쓰기 버전으로 열기
        File.write(toplevel_entry_type.get()) # 메모장에 입력한 내용 쓰기
        File.close() # 메모장 닫기
        toplevel_entry_type.delete(first=0, last=len(toplevel_entry_type.get())) #입력창 내용 삭제
        File = open("Students Name.txt", "r", encoding = "UTF-8") # 메모장 파일 읽기 위해 열기
        file_students_name = File.readline() # 사용자가 입력한 이름들 불러오기
        list_students_name = re.split(pattern='\s|,', string=file_students_name) # 정규표현식을 사용하여 , 공백으로 이름 리스트로 넣기
        File.close()
        list_students_name_final = [] # 학생 최종 이름 넣을 리스트 생성
        for i in list_students_name: # 학생 이름이 2글자 이상일 때만 제대로 된 이름. 해당 이름만 최종 리스트에 넣기       
            if len(i) >= 2:
                i = i.strip()
                list_students_name_final.append(i)
        toplevel_messagebox = tkinter.messagebox.askyesno(title="학생 이름 확인", message="아래의 학생 이름이 맞는지 확인하세요.\n\n{}".format(file_students_name)) #이름 맞는지 확인하기
        if toplevel_messagebox: # 이름 제대로 입력했다고 하면 학생 이름들 메모장에 넣기
            File = open("Students Name.txt", "w", encoding="utf-8")
            for i in list_students_name_final:
                File.write(i)
                File.write("\n")
            File.close()
            toplevel.destroy()  
            root.destroy()
        else: # 이름 제대로 입력 안되어있으면 다시 같은 과정 반복
            pass
    toplevel_entry_type.bind("<Return>", memo_write) # Enter만 쳐도 memo_write 함수 사용 되도록
    toplevel_button = tk.Button(toplevel, text="클릭", command=memo_write) # 버튼 클릭하면 memo_write 함수 사용 되도록
    toplevel_button.pack()
    root.mainloop()
    

#기본 시작 값
row_value=0 # 명단 밑에 넣을 라벨, 엔트리 입력을 위한 열 값 넣을 공간

# 기본 설정_앞
if os.path.isfile("Students Name.txt"): # Students Name 파일이 있다는 것은 이미 초기 세팅을 끝냈다는 것. 이름 등이랑 다 보이게 하고 초기 버튼 없어지게 하기
    root.title(string = "neis_students_record_by_raccoonTeacher") # 윈도우 제목 설정
    root.geometry("400x700+450+140") # 윈도우 크기와 좌표 설정 width=230, hegith=500, x=0, y=140
    root.resizable(width = 1, height = 1)

    #메뉴바 구성
    menu_students_name = tk.Menu(menubar) # 상위 메뉴 탭 항목 추가
    menubar.add_cascade(label="설정", menu = menu_students_name) # 상위 메뉴 탭 설정(보이게)
    menu_students_name.add_command(label="학생 명단 추가", command=execute_toplevel) # 항목 추가

    root.config(menu=menubar) #menubar 보이게

    File=open(file="Students Name.txt", mode="r", encoding="UTF-8") # 파일 열기
    for i, text in enumerate(File): # enumerate는 txt의 "인덱스, 문자"를 리스트로 반환해줌. 
        globals()["Name_{}".format(i)] = tk.Button(master=root, text=text) # var에 Cbutton_value_i 입력
        globals()["Name_{}".format(i)].grid(row=i//3, column=i%3) # 학생 이름 button 화면에 표시
        total_students_number = i
        row_value=i//3+2 # 명단 밑에 넣을 라벨, 엔트리 입력을 위한 열 값 확보
    File.close()
    root_label_time = tk.Label(text="[학생 행동 시간]\n\"09-01\"\n와 같은 방식으로 적기", background="#CCFFFF", width=30, pady=10).grid(row=row_value,column=1, columnspan=3)
    root_Entry_time = tk.Entry(master=root, width=30, border=5)
    root_Entry_time.grid(row=row_value+1,column=1, columnspan=3) # 학생 행동 시간 적는 곳
    root_label_content = tk.Label(text="[학생 행동]", background="#CCFFFF", width=30, pady=10).grid(row=row_value+2,column=1, columnspan=3)
    root_Entry_content = tk.Entry(master=root, width=30, border=5)
    root_Entry_content.grid(row=row_value+3,column=1, columnspan=3) # 학생 행동 적는 곳
    #root_Entry_content.bind("<Return>", insert_to_excel)

    #root_button_excel=tk.Button(master=root, text="엑셀에 저장하기", border=5, borderwidth=5, background="#FF99CC", command=insert_to_excel)
    #root_button_excel.grid(row=row_value+4,column=1, columnspan=3) # 엑셀에 저장하기 버튼
    #root_button_macro=tk.Button(master=root, text="나이스에 입력하기", border=5, borderwidth=5, background="#FF99CC", command=neis_macro)
    #root.bind("<Control-F1>", neis_macro)
    #root_button_macro.grid(row=row_value+5,column=1, columnspan=3) # 엑셀에 저장하기 버튼
    # 기본 설정_뒤
    root.mainloop()
else: # 해당 파일이 없다면 초기 세팅이 안되어있다는 것. 초기 세팅 하러 ㄲ\
    execute_toplevel()
```

```python
import tkinter as tk

# Function to read the names from the file
def read_names_from_file():
    names = []
    try:
        with open('StudentsName.txt', 'r') as f:
            for line in f:
                names.append(line.strip())
    except FileNotFoundError:
        # If the file does not exist, show an input dialog
        input_window = tk.Tk()
        input_window.title('Enter Names')
        input_label = tk.Label(input_window, text='Enter the names of the students, separated by newlines:')
        input_label.pack()
        input_text = tk.Text(input_window)
        input_text.pack()
        input_button = tk.Button(input_window, text='OK', command=input_window.destroy)
        input_button.pack()
        input_window.mainloop()
        # Split the input text by newlines and add each name to the list
        names = input_text.get('1.0', 'end').split('\n')
    return names

# Read the names from the file
names = read_names_from_file()

# Print the list of names
print(names)
```

```python
import tkinter as tk

# Create the main window
window = tk.Tk()

# Create a list of names
names = ['Alice', 'Bob', 'Charlie', 'Dave', 'Eve']

# Create a dictionary to store the counts for each person
counts = {name: 0 for name in names}

# Function to increment the count for a person
def increment_count(name):
    counts[name] += 1
    update_label(name)

# Function to update the label for a person
def update_label(name):
    label = labels[name]
    count = counts[name]
    label.config(text=str(count))

# Create a button and label for each person
buttons = {}
labels = {}
for name in names:
    button = tk.Button(text=name, command=lambda name=name: increment_count(name))
    button.pack()
    buttons[name] = button
    label = tk.Label(text='0')
    label.pack()
    labels[name] = label

# Create a reset button
reset_button = tk.Button(text='Reset', command=lambda: counts.update(zip(names, [0]*len(names))))
reset_button.pack()

# Run the main loop
window.mainloop()
```
